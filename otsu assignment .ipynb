```python
Histogram analysis 
You can consider histogram as a graph or plot,
which gives you an overall idea about the intensity distribution of an image.
It is a plot with pixel values (ranging from 0 to 255, not always) 
in X-axis and corresponding number of pixels in the image on Y-axis.
it is just another way of understanding the image. By looking at the histogram of an image,
you get intuition about contrast, brightness, intensity distribution etc of that image
(Remember, this histogram is drawn for  grayscale image, not color image)
** You can see the histogram for flower image ,left region of histogram shows the amount of darker pixels 
in image and right region shows the amount of brighter pixels.
From the histogram, you can see dark region is more than brighter region
** the histogram for swan image. 
left region of histogram shows the amount of bright pixels in image and the colours start to darker more 
than before then after this region shows brighter pixels an so on based on the intensity of the pixel value.
From the histogram, you can see dark region is more than brighter region.
** You can see the histogram for tools image .
green region of histogram shows the amount of darker pixels in image and white region shows the amount of brighter pixels.
From the histogram, you can see bright region is more than darker region.


```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img1 = cv2.imread('assignment1/flower.pgm',-1)
# global thresholding
ret1,th1 = cv2.threshold(img1,180,255,cv2.THRESH_BINARY)
# Otsu's thresholding
ret2,th2 = cv2.threshold(img1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)



for i in range(1):
   
    plt.subplot(1,3,1),plt.imshow(th1,'gray')
    plt.title('global threshold'), plt.xticks([]), plt.yticks([])
    plt.subplot(1,3,2),plt.imshow(th2,'gray')
    plt.title('otsu threshold'), plt.xticks([]), plt.yticks([])
    
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('assignment1/flower.pgm',0)
img = cv2.medianBlur(img,5)
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
cv2.THRESH_BINARY,11,2)
titles = ['Original ', 'Global Thresholding',
            'Adaptive Mean', 'Adaptive Gaussian']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
#the algorithm calculate the threshold for a small regions of the image.
#So we get different thresholds for different regions of the same image 
#and it gives us better results for images with varying illumination.
```
